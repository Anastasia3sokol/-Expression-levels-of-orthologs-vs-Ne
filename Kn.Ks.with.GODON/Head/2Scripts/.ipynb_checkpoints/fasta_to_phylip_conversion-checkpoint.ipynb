```python
import re
```

```python
files = !ls ../../Body/1_Raw/omm_NT_fasta.v10b_116tax_CDS_final/*
```

```python
files
```

```python
names = [re.sub('../../Body/1_Raw/omm_NT_fasta.v10b_116tax_CDS_final/', '', file) for file in files]
names = [re.sub('.fasta', '', file) for file in names]
```

```python
names
```

```python
for file, n in zip(files, names):
    f = open(file).readlines()
    new_f = []
    num = len([f[i] for i in range(0, len(f), 2)])
    length = len(f[1]) - 1
    new_f.append('  ' + str(num) + '   ' + str(length) + '\n')
    new_f.append('\n')
    for line in f:
        if line[0] == '>':
            new_f.append(line[1:])
        else:
            for i in range(len(line.strip())//59 + 1):
                new_f.append(line.strip()[i*59:59*(i+1)] + '\n')
            new_f.append('\n')
            
    name = '../../Body/1_Raw/omm_NT_phylip.v10b_116tax_CDS_final/' + n + '.phylip'
    with open(name, 'w') as phy:
        for line in new_f:
            phy.write(line)
```
